## (Aside - discussed earlier) Modeling Framework: SATIMGE

The ESRG's SATIMGE modeling framework comprises two core components:

1. **SATIM (South African TIMES Model)**: A comprehensive whole-system energy model designed to optimize energy pathways while adhering to technical, environmental, and economic constraints. SATIM employs a bottom-up approach to energy system optimization, meticulously dissecting South Africa's energy sector into supply and consumption domains. The supply side encompasses electricity generation, liquid fuels refining, and primary energy inputs, while the consumption side encompasses residential, commercial, industrial, and transport sectors. Each demand sector is further stratified into subsectors, such as iron and steel or passenger transport, culminating in a granular breakdown at the technology level. These technologies are characterized by parameters including fuel use, cost, efficiency, and emissions, among others. When executing a scenario, SATIM forecasts annual energy demand based on macroeconomic drivers like GDP and population growth, subsequently determining the least-cost technology mix to satisfy this demand over the modeling horizon, while adhering to specified constraints like emissions limits and technological availability.

2. **SAGE (South African General Equilibrium Model)**: SAGE is a dynamic, recursive, economy-wide model that simulates the intricate workings of the South African economy. With a comprehensive representation encompassing up to 104 sectors and commodities, 5 factors of production, and 12 household categories, SAGE facilitates in-depth analysis of economic dynamics. Within each annual time slice, SAGE iteratively solves based on population, productivity, and capital supply data and assumptions. Between periods, the model updates to reflect changes such as population growth, technological advancements, and capital accumulation. New capital formation is endogenously determined based on previous-period investment levels and sector profitability, thereby ensuring sector-specific capital allocation. While SAGE possesses limited capacity to internally assimilate technological advancements, particularly for extreme long-term scenarios, it guarantees technical viability within forecasted bounds.

## Integrated Model

The energy-extended version of SAGE interfaces seamlessly with the comprehensive energy sector modeling of SATIM. This integration enables a holistic assessment of energy system changes on various economic sectors, markets, and agents. Through iterative interaction, information flows bidirectionally between SATIM and SAGE:

1. **SATIM computes the least-cost energy technology mix**, incorporating assumptions regarding technology and fuel costs, as well as constraints like demand and emissions targets.
2. The resulting technology mix is transmitted to **SAGE**, which assimilates the new energy supply and demand composition into its economic projections. Aggregate investment in SAGE is determined via a macroeconomic adjustment process, wherein changes in aggregate final demand are apportioned proportionally across consumption, investment, and government spending.
3. SAGE furnishes updated economic indicators, which are then fed back into **SATIM** for demand refinement and further optimization.
4. This iterative exchange continues until convergence is achieved in terms of energy utilization, CO2 emissions, demand, price, and technology mix.



## 5.1 Demand Modeling: Long-term Energy Demand Analysis

### 5.1.1 Definition of Energy Demand

The term 'energy demand' is defined herein as the consumption of a final energy carrier, such as electricity or petrol, for specific energy end-uses, such as lighting in residential homes or driving a vehicle. This metric is typically measured in units of energy, such as Peta-Joules (PJ) or TeraWatt-hours (TWh) for electricity consumption.

It's important to note that this definition differs from Eskom's interpretation of 'energy demand', where it refers to the amount of power drawn from the grid at a given moment, measured in MegaWatts (MW).

### 5.1.2 Drivers of Energy Demand

The principal drivers of energy demand growth in SATIM are population and GDP. Population growth directly impacts energy service demand in residential and transport sectors, and indirectly affects all other sectors. GDP growth influences energy service demand across the model, as it correlates with industrial production, commercial sector activity, transport, and household income distribution.

## 5.2 Sectoral Energy Demand

### 5.2.1 Residential Buildings

The residential sector in SATIM is structured to capture changes in fuel and technology use alongside energy service demands resulting from rising income levels. It comprises three income groups:

- Low income (~45% of the population)
- Middle income (~31% of the population)
- High income (~24% of the population)

Each income group is further segmented to reflect demand for various energy services, including lighting, cooking, space heating, refrigeration, and other electrical appliances. Household classification and primary fuel assumptions are derived primarily from the 2011 Census and the National Development Plan.

Fuel stacking, wherein households use multiple fuel types to meet energy service needs, is common, especially among low and middle-income households. Lower-income groups predominantly use liquid (paraffin) and solid fuels (coal and wood), with lower demand for electricity due to appliance ownership limitations.

SATIM models household energy service needs met by specific technologies and fuels, with assumed average efficiencies. This allows for the assessment of energy demand impacts resulting from shifts in fuel or technology use. Technological efficiencies can vary across income groups, reflecting differences in appliance quality and efficiency levels.

The upward mobility of households from lower to higher income groups, driven by GDP growth, is anticipated to increase electricity usage over time as households expand appliance ownership. However, not all households may transition to higher income brackets by 2050, necessitating ongoing support for those in energy poverty.

## 5.2.2 Commercial Sector

**Sector Composition**: 
  - The commercial buildings sector in SATIM encompasses:
    - Wholesale, retail, and motor trade services
    - Warehouses and industrial spaces
    - Office buildings
    - Banking facilities
    - Other non-residential and government spaces including schools, hospitals, hotels, and recreational facilities.
  - Public sector floor area serving public services, including public lighting and water services, is also considered.

**Energy Demand Estimation**:
  - Commercial sector energy service demand is estimated based on the energy intensity of energy	services required as a function of lettable floor area (PJ/m2); the growth in floor area over time and improvements in energy efficiency as a result of the implementation of successive building code regulations.
  -  Final energy demand is internally computed within SATIM, factoring in technology efficiency and penetration within the sector.
  - One of the most significant drivers in the commercial buildings sector is the total commercial floor area, which draws from Statistics South Africa's data on floor area additions. Growth in floor area over the model horizon is informed by the historical correlation between floor area growth and GDP growth, with a lag of approximately two years observed between changes in floor area and real economic growth. New building floor area projections consider the rate of commercial floor area growth and the loss in existing floor area via demolitions.
  - The model assumes that energy services in new buildings can be delivered more efficiently with newer technologies, while constraints on residual capacity limit the adoption of newer technologies in older buildings

**Energy Service Assessment**:
  - Energy services are assessed via a bottom-up approach, focusing on highly energy-intensive services such as:
    - Cooling, space heating, cooking, lighting, refrigeration, water heating, public lighting, and public water supply.
  - Less energy-intensive services are aggregated under 'other'.
  - The model provides detailed insights into technology type and fuel choices for supplying these energy services.
  - End-uses can be represented either by fuel or technology. Fuel shares indicate significant energy services within sub-sectors and their respective proportions of energy demand. Technology shares encompass the prevalence, activity levels, and efficiencies of different technology types supplying specific energy services. 

#### The Sector's Potential Future Energy and Fuel Use
The commercial buildings sector in South Africa is positioned to undergo significant transformations driven by sustainability goals, equitable urban planning initiatives, and technological advancements. Sustainable and energy-efficient designs may dominate, incorporating passive architectural design elements that prioritise the use of ambient energy such as advanced insulation, high performance windows, airtightness, natural ventilation and daylighting systems with heat recovery, elimination of thermal bridges and other thermally efficient building envelope elements are poised to reshape building materials and drastically reduce the need for extensive heating and cooling. Access to essential utilities like heating, cooling, and electricity will likely rely heavily on electrified, energy-efficient and sustainable solutions, reflecting enhanced environmental responsibility. In terms of specific systems, there will likely be a notable shift towards electric-based technologies for heating, cooking, water heating, and lighting, with an increasing reliance on renewable energy sources.  

Water heating systems may undergo a transformative shift, marked by a surge in the integration of solar technologies and efficient electric water heaters, optimising hot water generation through renewables. Highly efficient LED and fluorescent lighting will likely become the second most widely used lighting technologies and the share of high-intensity discharge (HID) and incandescent lights in existing buildings will decline, indicative of retrofitting efforts. We anticipate the widespread adoption of smart lighting systems, featuring sensors and automation to optimise illumination based on occupancy and natural light availability, leading to substantial energy savings. Passive daylighting strategies will also gain prominence, involving the use of static systems like windows, skylights, light tubes, mirrors, and light shelves to capture and reflect light into key areas of the building. 

Changing occupancy patterns, including the adoption of flexible work arrangements and telecommuting, would also influence the demand for heating and cooling systems during traditional office hours. Occupants' behaviour, guided by heightened energy conservation awareness, will likely further encourage responsible heating and cooling system usage.  

Regulatory frameworks are expected to evolve, emphasising optimisation and the integration of smart grid-type devices, which may be overseen by electricity grid operators. These developments would play a pivotal role in managing energy variability, enhancing storage, and optimising building energy use while aligning with sustainability objectives. Moreover, inclusive urbanisation strategies may address spatial inequalities by encouraging mixed-use developments and fostering social cohesion in urban planning initiatives. Additionally, we foresee an increased adherence to green building codes such as SANS10400 XA2, accompanied by the introduction of energy endorsement labels for energy-efficient appliances. Minimum Energy Performance Standards (MEPS) should also be rigorously implemented to further promote energy efficiency and sustainability across the sector.

## 5.2.4 The Transport Sector

Passenger demand is quantified using the ‘passenger-kilometre’ (pkm) metric, considering various parameters such as income level, private vehicle ownership, vehicle occupancy, annual mileage, and travel time budget. SATIM utilizes economic drivers like GDP and population growth to project future passenger transport demand by vehicle type. Subsequently, the model determines the optimal fuel type for new vehicle purchases each year to meet this demand, thus updating the country's vehicle stock accordingly.

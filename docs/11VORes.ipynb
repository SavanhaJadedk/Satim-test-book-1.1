## 11.1 Modelling Structure

*Table 11.1: Modelling Structure and Disaggregation for the Residential Sector in SATIM*

| **Parameter**          | **Disaggregation Level**                                     | **Drivers**                                                                                                                        |
|------------------------|--------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------|
| **Households**          | Low, Middle, and High-income groups                          | - **Population Growth**: Influences the total number of households.<br> - **Household Size**: Variations in the number of people per household affect total energy consumption.<br> - **GDP**: Economic growth impacts household income, driving energy demand and appliance ownership. |
| **Energy Service Demands** | Lighting, cooking, water heating, space heating and cooling, refrigeration, other appliances | - **Household Income**: Determines the ability of households to afford modern energy services and appliances.<br> - **Electrification**: Access to electricity allows for a transition from traditional fuels (e.g., wood, paraffin) to modern fuels (e.g., electricity, gas).<br> - **Policies & Regulations**: Government regulations such as building energy codes, minimum energy performance standards (MEPS), and appliance standards influence energy efficiency and consumption.<br> - **Behavioral Changes**: Shifts in consumer behavior, such as energy conservation or the adoption of efficient technologies (e.g., LED lighting), affect energy demand patterns. |
| **Fuels and Appliances** | Wood, coal, paraffin, gas, electricity (appliances like televisions, washing machines, dishwashers, etc.) | - **Household Income**: Higher incomes enable the purchase of modern appliances and the transition to cleaner fuels.<br> - **Electrification**: As electrification expands, households shift from traditional fuels (biomass, paraffin) to electricity-based appliances.<br> - **Policies & Subsidies**: Government initiatives such as fuel subsidies and appliance standards encourage the use of more efficient technologies and cleaner energy sources.<br> - **Appliance Standards**: The enforcement of standards like MEPS drives the adoption of energy-efficient appliances, reducing energy consumption per household. |


## 11.2 Navigating the Residential Sector with Veda Online


*Table 11.2: Attributes used in the Residential Sector*


| **Attribute**        | **Description**                                           | **Purpose in Model**                                                 | **Units**            |
|----------------------|-----------------------------------------------------------|------------------------------------------------------------|----------------------|
| **TechName**         | Name of the technology or appliance (e.g., Electric Heater). | Identifies the technology being modeled.                   | N/A                  |
| **Comm-IN**          | Incoming commodities or inputs to the technology.        | Represents the resources entering the technology.          | N/A   |
| **Comm-OUT**         | Outgoing commodities or outputs from the technology.     | Represents the products generated by the technology.       | N/A  |
| **EFF~(year)**              | Efficiency of the technology or activity in converting energy.       | Indicates how effectively the technology utilizes energy, for a given year.   | Percentage (%)        |
| **STOCK**            | Stock of the technology available in the residential sector. | Reflects the total amount of existing installed capacity of the technology.       | Petajoule-annum (PJa)|
| **INVCOST**          | Investment cost associated with the technology.          | Represents the capital investment required for the technology. | 2005 mill. R/PJ-a   |
| **NCAP_BND~0**       | Capacity bound at the initial period (year 0).           | Sets the initial capacity limits for the technology.       | kW or MW             |
| **FLO_MARK (UP)**    | Upper bound on the flow of energy between processes or from a process to a market. | Sets an upper constraint on how much energy (or material) can flow from a technology (e.g., solar heater) to a market. | Percentage (%)                  |
| **FLO_MARK (LO)**    | Lower bound on the flow of energy between processes or from a process to a market. | Sets a minimum requirement for how much energy must flow to a market from a given technology (e.g., electric heaters). | Percentage (%)                  |

## Data Sources and Assumptions

- **Household Income Classification**: Based on the 2011 Census (StatsSA, 2011).
- **Electrification**: Assumed according to the National Development Plan (National Planning Commission, 2012).
- **Efficiency Improvements**: Driven by the Standards and Labelling Programme and MEPS (DMRE, 2022), and National Energy Efficiency Standards (NEES), 2015.

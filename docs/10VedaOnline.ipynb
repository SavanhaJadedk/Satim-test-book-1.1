# 10. Veda Online


## 10.1 Navigating Veda Online to Retrieve SATIMGE Input Data

Accessing SATIMGE data on Veda Online allows users to filter by scenarios, processes, commodities, attributes, and user constraints. Follow these steps to efficiently retrieve the specific input data for the SATIMGE model.

#### Accessing Veda Online
- Navigate to the Veda Online platform: [https://vedaonline.cloud/Home.aspx](https://vedaonline.cloud/Home.aspx).
- In the left panel, look for **public models** and select **brunom’s models**.
- Choose the **SATIMGE** model to begin.
- SATIMGE Veda Online quick link: [https://vedaonline.cloud/ModelPage.aspx]( https://vedaonline.cloud/ModelPage.aspx) 

#### Using the "Browse" Tab
- Once inside the SATIMGE model, click on the **Browse** tab located in the top navigation bar. 
- From here, you will have access to several filters, allowing you to narrow down your data search by:
  - **Scenario**
  - **Process**
  - **Commodity**
  - **Attribute**
  - **User Constraint**

#### Filtering by Process
You can filter data by sector or subsector using the **Process** filter, which includes the following:
- **Main Demand Sectors**:
  - Residential
  - Commerce
  - Transport
  - Industry
- **Other Sectors and Subsectors**:
  - Imports
  - Exports
  - Agriculture
  - Power

#### Filtering by Commodity
The **Commodity** filter enables users to search for commodities relevant to different sectors:
- **Residential Commodities**: Use the wildcard `R*` to search for all residential commodities.
- **Commercial Commodities**: Search with `C*` to retrieve all commodities related to the commercial sector.
- **Transport Commodities**: Use `T*` to find transport sector-related commodities.
- For specific commodities, use the exact commodity code or refer to the breakdown provided for each sector in this documentation.

#### Filtering by Attributes and User Constraints
In addition to filtering by process and commodity, users can filter by **Attributes** or **User Constraints**:
- **Attributes**: This includes parameters like **ACT_COST**, **FLO_MARK**, **EFF**, and others specific to each sector.
- **User Constraints**: These allow users to apply specific limits or requirements to the model inputs.

#### Example Filters and Selected Parameters
For each sector, this documentation provides an example of filters and selected parameters, guiding users on how to filter data efficiently. These examples can be used to locate specific model inputs or constraints for your analysis.

#### Exporting or Downloading Data
- Once you’ve applied the necessary filters, you can download the model version and scenario-specific data:
  - **Excel file** or **CSV** format.
  - Alternatively, Veda Online provides a link to the model's **GitHub** repository, where users can access the actual input files directly.
